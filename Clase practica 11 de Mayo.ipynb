```sas
proc iml;
    * muestras de una mezcla gaussiana;
    pi = {0.25 0.5 0.25}; 
    mu1 = -{3,2};
    mu2 = {0,0};
    mu3 = {2,3};
    
    V1 = {1 -0.4, -0.4 1};
    V2 = {1 0, 0 1};
    V3 = {1.5 0.7, 0.7 1.5};
    
    clus = RandMultinomial(1, 2000, pi);
    
    
    *seleccionamos elatoriamemte el tamano de los clusters;
    n1 = clus[1];
    n2 = clus[2];
    n3 = clus[3];
    
    cluster = j(n1,1,1)//j(n2,1,2)//j(n3,1,3); 
    
    print clus;
    
    rnd1 = RandNormal(n1, mu1, V1);
    rnd2 = RandNormal(n2, mu2, V2);
    rnd3 = RandNormal(n3, mu3, V3);
    
    *agregamos el que indica el cluster al cual pertenece la muestra;
    sample = rnd1 // rnd2 //rnd3; *concatenar verticalmente las muestras;
    
    x = sample[,1];
    y = sample[,2];
        
    *creacion de la tabla;
    create simulacion_cluster var{x y cluster}; append;
    close simulacion_cluster;
    
quit;
```
