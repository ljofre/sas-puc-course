sas tiene escencialmente dos tipos de instruccionaes: **data** y **proc**

```sas
DATA PACTIVO;
INPUT NUM_PAC TRATAM INIC $ SEXO EDAD;
CARDS;
1 1 JPP 1 53
2 2 JDA 1 47
3 1 RRA 0 81
4 2 HSB 0 29
;
RUN;
```

```sas
proc print;
run;
```

```sas
* guardamos la data en un archivo de texto;

proc export data = work.pactivo
    outfile = '/folders/myfolders/tutorial/pactivo.txt'
    dbms=tab replace;
    putnames=yes;
run;
```
